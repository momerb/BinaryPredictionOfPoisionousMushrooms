```python
def filter_by_value_counts(dataframe, column, min_value):
    
    value_counts = dataframe[column].value_counts()

    return dataframe[dataframe[column].isin(value_counts[value_counts >= min_value].index)]
```

```python
def one_hot_encode_with_drop(dataframe, columns):

    df_final = dataframe.copy()
    
    for col in columns:

        dummies = pd.get_dummies(dataframe[col], prefix=col)
        
        dummies = dummies.iloc[:, :-1]

        df_final = df_final.drop(col, axis=1)

        df_final = pd.concat([df_final, dummies], axis=1)
    
    return df_final
```
