```python
columns_float = ['cap-diameter', 'stem-height', 'stem-width']

columns_str = [col for col in df.columns.tolist() if col not in (['class', 'id'] + columns_float)]

columns_str, columns_float
```

```python
def one_hot_encode_with_drop(dataframe, columns):

    df_final = dataframe.copy()
    
    for col in columns:

        dummies = pd.get_dummies(dataframe[col], prefix=col)
        
        dummies = dummies.iloc[:, :-1]

        df_final = df_final.drop(col, axis=1)

        df_final = pd.concat([df_final, dummies], axis=1)
    
    return df_final
```

```python
df_encoded = one_hot_encode_with_drop(dataframe=df, columns=columns_str)

df_encoded
```
